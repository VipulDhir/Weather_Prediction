```python
# we r going to predict tomorrows maximum temp
# shift negative 1 will pull evry row back by one position
core_weather["target"] = core_weather.shift(-1)["temp_max"]
core_weather
```

```python
# To remove last row
# it is nan because we dont have the data for 19 aug so it cant predict
core_weather = core_weather.iloc[:-1,:].copy()
```

```python
core_weather
```

```python
from sklearn.linear_model import Ridge

reg = Ridge(alpha=.1)
```

```python
# what variables we are using to predict the temperature so we created a list
predictors = ["precip", "temp_max", "temp_min"]

train = core_weather.loc[:"2020-12-31"]
test = core_weather.loc["2021-01-01":]
```

```python
# calling fit mehtod and fit it to our training dataset
reg.fit(train[predictors], train["target"])
```

```python
# we will generate predictions on our test data using the predictor cols
predictions = reg.predict(test[predictors])
```

```python
from sklearn.metrics import mean_absolute_error
# mean absolute will subtract the actual from the predictions take the absolute value and find the average of that across all of the predictions
mean_absolute_error(test["target"], predictions)

# this means on an avg we are about 3.2 degrees off the actual temperature
```

```python
# we need to combined the two cols so that we can see the actual diff
# but predictions is in the numpy format we need to convert it into pandas series
combined =pd.concat([test["target"], pd.Series(predictions ,index = test.index)],axis=1)
combined.columns = ["actual", "predictions"]
combined
```

```python
combined.plot()
# our predictions are a bit more conservative
```

```python
# taking a look at our coefficients
reg.coef_
#precipation has negative impact on temperature
```

```python
def create_predictions(predictors, core_weather, reg):
    train = core_weather.loc[:"2020-12-31"]
    test = core_weather.loc["2021-01-01":]

    reg.fit(train[predictors], train["target"])
    predictions = reg.predict(test[predictors])

    error = mean_absolute_error(test["target"], predictions)

    combined = pd.concat([test["target"], pd.Series(predictions, index=test.index)], axis=1)
    combined.columns = ["actual", "predictions"]
    return error, combined
```

```python
# finding avg of the month
# for every row it will calculate avg of previous 30 days
core_weather["month_max"] = core_weather["temp_max"].rolling(30).mean()
core_weather
```

```python
# for a given day was the temperature different than the month monthly mean
core_weather["month_day_max"] = core_weather["month_max"] / core_weather["temp_max"]
core_weather
```

```python
# ratio between max and min temperatures
core_weather["max_min"] = core_weather["temp_max"] / core_weather["temp_min"]
core_weather
```

```python
predictors = ["precip", "temp_max", "temp_min", "month_day_max", "max_min"]
```

```python
core_weather = core_weather.iloc[30:,:].copy()
```

```python
error, combined = create_predictions(predictors, core_weather, reg)
error
```
